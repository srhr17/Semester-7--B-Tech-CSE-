```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
%matplotlib inline
```

```
iris=sns.load_dataset('iris')
```

```
iris.head()
```

```
sns.jointplot(x=iris.sepal_length,y=iris.sepal_width)
```

```
sns.FacetGrid(iris,hue='species',size=5)\
.map(plt.scatter,'sepal_length','sepal_width')\
.add_legend()
```

```
sns.boxplot(x='species',y='petal_length',data=iris)
```

```
sns.stripplot(x='species',y='sepal_length',data=iris,jitter=True,edgecolor='gray')
```

```
ax=sns.boxplot(x='species',y='sepal_length',data=iris)
ax=sns.stripplot(x='species',y='sepal_length',data=iris,jitter=True,edgecolor='gray')
```

```
sns.violinplot(x='species',y='sepal_length',data=iris)
```

```
sns.pairplot(data=iris,kind='scatter')
```

```
sns.pairplot(iris,hue='species')
```

```
cor=iris.corr()
sns.heatmap(cor,annot=True,cmap='summer')
```

```
iris.hist(edgecolor='black', linewidth=1.2)
fig=plt.gcf()
fig.set_size_inches(12,6)
plt.show()
```

```
sns.set(style="whitegrid")
fig=plt.gcf()
fig.set_size_inches(10,7)
fig = sns.swarmplot(x="species", y="petal_length", data=iris)
```
